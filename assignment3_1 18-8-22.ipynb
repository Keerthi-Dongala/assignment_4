```python
df= plt.figure()
a= df.add_axes([0,0,1,1])
a.plot(x,y)
a.set_xlabel('x')
a.set_ylabel('y')
a.set_title('title')
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
df, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))
axes[0].plot(x,y,color="blue", linewidth=3)
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
df
axes[1].plot(x,z,color="red", linewidth=2, linestyle="dashed")
axes[1].set_xlabel('x')
axes[1].set_ylabel('z')
df
```
