```python
import matplotlib.pyplot as plt
country=['USA','UK','INDIA','KOREA']
GDP_PER_CAPITA=[7400,6900,7500,6400]
plt.bar(country,GDP_PER_CAPITA)
plt.title('country vs per capita')
plt.xlabel("country")
plt.ylabel("GDP_PER_CAPITA")
plt.show()
```
