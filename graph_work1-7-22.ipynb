```python
import matplotlib.pyplot as plt
import numpy as np
plt.title("GRAPH")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
xpoints=np.array([0,0,2,2,0,2,0,2,1])
ypoints=np.array([20,0,0,20,20,0,0,20,10])
plt.plot(xpoints,ypoints)
plt.plot(xpoints,ypoints,marker="o",color="b",mec="r")
plt.show()
```
