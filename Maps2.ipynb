```python
#resize our map by using the branca library in python
from branca.element import Figure
fig=Figure(width=500,height=500)
map1=folium.Map(location=[17.3573,82.5443],zoom_start=14,min_zoom=10,max_zoom=16)
fig.add_child(map1)
map1
```

```python
fig2=Figure(width=500,height=500)
map2=folium.Map([17.3573,82.5443])
fig2.add_child(map2)
folium.TileLayer('Stamen Terrain').add_to(map2)
folium.TileLayer('Stamen Toner').add_to(map2)
folium.TileLayer('Stamen Water Color').add_to(map2)
folium.TileLayer('Cartodbpositron').add_to(map2)
folium.TileLayer('Cartodbdark_matter').add_to(map2)
folium.LayerControl().add_to(map2)
map2
```

```python
#creating basemap
fig3=Figure(width=800,height=500)
map3=folium.Map(location=[17.3573, 82.5443],tiles='cartodbpositron',zoom_start=8)
fig3.add_child(map3)
#adding markers to the map
folium.Marker(location=[17.3573,82.5443],popup='Default pop up marker1',tooltip="click here to see more").add_to(map3)
folium.Marker(location=[17.5472,82.8573],popup='<strong>marker2</strong>',tooltip='<strong>click here to see more</strong').add_to(map3)
folium.Marker(location=[17.6896,83.0024],popup='<h3 style="color:blue">marker3</h3',tooltip="click here to see more").add_to(map3)
folium.Marker(location=[17.6868,83.2185],popup='<h4>marker4</h4>',tooltip="click here").add_to(map3)
map3
```
